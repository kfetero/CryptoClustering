```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_crypto_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv", index_col="coin_id"
)

# Display sample data
df_crypto_market_data.head(10)
```

```python
# Generate summary statistics
df_crypto_market_data.describe()
```

---

```python
# Plot your data to see what's in your DataFrame
df_crypto_market_data.hvplot.line(width=800, height=400, rot=90)
```

### Prepare the Data

```python
df_crypto_market_data.columns
```

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
crypto_market_data_scaled = StandardScaler().fit_transform(
    df_crypto_market_data[
        [
            "price_change_percentage_24h",
            "price_change_percentage_7d",
            "price_change_percentage_14d",
            "price_change_percentage_30d",
            "price_change_percentage_60d",
            "price_change_percentage_200d",
            "price_change_percentage_1y",
        ]
    ]
)
```

```python
# Create a DataFrame with the scaled data
df_crypto_market_scaled = pd.DataFrame(
    crypto_market_data_scaled,
    columns=[
        "price_change_percentage_24h",
        "price_change_percentage_7d",
        "price_change_percentage_14d",
        "price_change_percentage_30d",
        "price_change_percentage_60d",
        "price_change_percentage_200d",
        "price_change_percentage_1y",
    ],
)

# Copy the crypto names from the original data
df_crypto_market_scaled["coin_id"] = df_crypto_market_data.index
# Set the coin_id column as index
df_crypto_market_scaled = df_crypto_market_scaled.set_index("coin_id")

# Display sample data
df_crypto_market_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_list = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`(df_crypto_market_scaled)
# 3. Append the model.inertia_ to the inertia list
for i in k_values:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_crypto_market_scaled)
    inertia_list.append(k_model.inertia_)
    
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia_list}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
elbow_df.head(10)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
original_data_elbow_plot = elbow_df.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k_values)

# show the plot
original_data_elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k is 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data df_crypto_market_scaled
kmeans_model.fit(df_crypto_market_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_market_data_scaled_clusters = kmeans_model.predict(df_crypto_market_scaled)

# Print the resulting array of cluster values.
print(crypto_market_data_scaled_clusters)
```

```python
# Create a copy of the DataFrame
df_crypto_market_scaled_predictions = df_crypto_market_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_crypto_market_scaled_predictions[
    "crypto_cluster"
] = crypto_market_data_scaled_clusters

# Display sample data
df_crypto_market_scaled_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Colour the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
clusters_original_data_plot = df_crypto_market_scaled_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="crypto_cluster",
    hover_cols="coin_id",
)
# Show the plot
clusters_original_data_plot
```

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to
# three principal components.
crypto_info_pca = pca_model.fit_transform(df_crypto_market_scaled_predictions)
# View the first five rows of the DataFrame.
crypto_info_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
pca_model.explained_variance_ratio_
```

```python
# Create a new DataFrame with the PCA data
df_pca_data = pd.DataFrame(crypto_info_pca, columns=["PCA1", "PCA2", "PCA3"])

# Copy the crypto names from the original data
df_pca_data["coin_id"] = df_crypto_market_scaled_predictions.index

# Set the coin_id column as index
df_pca_data = df_pca_data.set_index("coin_id")

# Display sample data
df_pca_data.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values_pca = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_list_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k_values_pca:
    k_model_pca = KMeans(n_clusters=i, random_state=0)
    k_model_pca.fit(df_pca_data)
    inertia_list_pca.append(k_model_pca.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {"k": k_values_pca, "inertia": inertia_list_pca}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_data)
pca_elbow_df.head(10)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
pca_elbow_plot = pca_elbow_df.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k_values_pca)

#show the plot
pca_elbow_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k using the PCA data is 4 as the Elbow Curve shows.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** It does not differ from the best k value found using the original data. 
  Based on the above Elbow Curve, it looks like `k=4` is still the correct one.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k
pca_kmeans_model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
pca_kmeans_model.fit(df_pca_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
clusters_predict_pca = pca_kmeans_model.predict(df_pca_data)
# Print the resulting array of cluster values.
clusters_predict_pca
```

```python
# Create a copy of the DataFrame with the PCA data
pca_predictions_df = df_pca_data.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_predictions_df["predict_clusters"] = clusters_predict_pca

# Display sample data
pca_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Colour the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
clusters_pca_data_plot = pca_predictions_df.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="predict_clusters",
    hover_cols="coin_id",
)
# show the plot
clusters_pca_data_plot
```
